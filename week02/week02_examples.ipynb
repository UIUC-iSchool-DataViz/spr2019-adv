```python
@ipywidgets.interact(little_x_origin = (0.0, 15.0, 0.1), little_y_origin = (0.0, 15.0, 0.1),
                     little_height = (0.1, 2.0, 0.1), big_height = (1.0, 15.0, 0.1),
                     col = ipywidgets.ColorPicker(value="black"))
def make_boxes(little_x_origin = 0.0, little_y_origin = 0.0,
               little_height = 1.0, big_height = 10.0, col = "black"):
    fig = plt.figure(figsize=(6,6))
    ax1 = fig.add_axes([0.0, 0.0, 1.0, 1.0])

    big_width = area1 / big_height

    rect1 = pt.Rectangle( (0.0, 0.0), big_width, big_height, facecolor = col, edgecolor = "black" )
    ax1.add_patch(rect1)
    ax1.set_xlim(0, 15)
    ax1.set_ylim(0, 15)

    rect2 = pt.Rectangle( (little_x_origin, little_y_origin), area2 / little_height, little_height, facecolor = "red",
                         edgecolor = "black")
    ax1.add_patch(rect2)
    return fig
```
